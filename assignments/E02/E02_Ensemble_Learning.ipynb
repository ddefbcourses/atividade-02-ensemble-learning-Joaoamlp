# Questão 1

Implemente uma função load_data(seed) que:

Carregue o dataset `Fashion MNIST`
Realize a separação em treino e teste
Utilize `train_test_split` com controle de aleatoriedade
Retorne: `X_train`, `X_test`, `y_train`, `y_test`

Depois responda: 
É necessário normalizar os dados para esse tipo de modelo? Justifique.

**Solução**:

````python
import matplotlib.pyplot as plt


# TODO: implemente load_data

def load_data() :
    from sklearn.datasets import fetch_openml
    from sklearn.model_selection import train_test_split

    X, y = fetch_openml('Fashion-MNIST', version=1, as_frame=False, return_X_y=True)
        
    y = y.astype(int)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, 
        test_size=10000, 
        stratify=y  
    )

    X_train = X_train / 255.0
    X_test = X_test / 255.0

    return X_train, X_test, y_train, y_test

   
````

### Resposta 

Sim, é recomendado a normalização como estamos lidando com imagens o computador não tem a visão geral ele recebe valores numericos que representam os pixeis de 0 a 255 então como os valores podem chegar a varia de maneira brusca. É interessante ocorrer a normalização tanto pela velocidade do modelo quanto pela chance dele acertar.

````python
# TODO: implemente load_data



def random_forest(seed, x_train, y_train  ) :
    from sklearn.ensemble import RandomForestClassifier

    modelo_random_forest = RandomForestClassifier(n_estimators= 100, random_state= seed, n_jobs= 1)

    modelo_random_forest.fit(x_train,y_train)

    return modelo_random_forest

def ada(seed, x_train, y_train) :
    from sklearn.ensemble import AdaBoostClassifier
    
    modelo_adaboost = AdaBoostClassifier(n_estimators=100, random_state=seed)

    modelo_adaboost.fit(x_train, y_train)

    return modelo_adaboost


````

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados

- Avaliar o modelo
- Retornar a acurácia

**Solução**:

````python
# TODO: implemente load_data

def run_pipeline(model_type, seed):
    acuracia = 0

    X_train, X_test, y_train, y_test = load_data()

    if(model_type == "rf"):

        modelo = random_forest(seed, X_train,y_train)

        
    elif(model_type == "ad"):

        modelo = ada(seed, X_train,y_train)

        
    acuracia = evaluate(modelo,X_test,y_test)

    print(acuracia)
       
    return acuracia
def check_overfitting(X_train, X_test, y_train, y_test, seed=42):
    from sklearn.ensemble import RandomForestClassifier

    train_scores = []
    test_scores = []
    n_estimators_range = [1, 5, 10, 20, 50, 100] # "Camadas" de complexidade

    for n in n_estimators_range:
        # Criamos o modelo com 'n' árvores
        model = RandomForestClassifier(n_estimators=n, random_state=seed, n_jobs=-1)
        model.fit(X_train, y_train)
        
        # Avaliamos nos dois conjuntos
        train_scores.append(model.score(X_train, y_train))
        test_scores.append(model.score(X_test, y_test))
        
    return n_estimators_range, train_scores, test_scores

X_train, X_test, y_train, y_test = load_data()

n_range, train_acc, test_acc = check_overfitting(X_train, X_test, y_train, y_test)

plt.plot(n_range, train_acc, label='Acurácia Treino', marker='o')
plt.plot(n_range, test_acc, label='Acurácia Teste', marker='o')
plt.xlabel('Número de Estimadores (Complexidade)')
plt.ylabel('Acurácia')
plt.title('Identificando o Ponto de Overfitting')
plt.legend()
plt.show()
````

# Questão 1

Implemente uma função load_data(seed) que:

Carregue o dataset `Fashion MNIST`
Realize a separação em treino e teste
Utilize `train_test_split` com controle de aleatoriedade
Retorne: `X_train`, `X_test`, `y_train`, `y_test`

Depois responda: 
É necessário normalizar os dados para esse tipo de modelo? Justifique.

**Solução**:

````python
import matplotlib.pyplot as plt


# TODO: implemente load_data

def load_data() :
    from sklearn.datasets import fetch_openml
    from sklearn.model_selection import train_test_split

    X, y = fetch_openml('Fashion-MNIST', version=1, as_frame=False, return_X_y=True)
        
    y = y.astype(int)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, 
        test_size=10000, 
        stratify=y  
    )

    X_train = X_train / 255.0
    X_test = X_test / 255.0

    return X_train, X_test, y_train, y_test

   
````

### Resposta 

Sim, é recomendado a normalização como estamos lidando com imagens o computador não tem a visão geral ele recebe valores numericos que representam os pixeis de 0 a 255 então como os valores podem chegar a varia de maneira brusca. É interessante ocorrer a normalização tanto pela velocidade do modelo quanto pela chance dele acertar. 


# Questão 2

Implemente as funções:

`train_random_forest(X_train, y_train, seed)`
`train_adaboost(X_train, y_train, seed)`

## Requisitos:

Utilizar os modelos do `sklearn`
Garantir reprodutibilidade com `random_state`

**Solução**:

````python
# TODO: implemente load_data



def random_forest(seed, x_train, y_train  ) :
    from sklearn.ensemble import RandomForestClassifier

    modelo_random_forest = RandomForestClassifier(n_estimators= 100, random_state= seed, n_jobs= 1)

    modelo_random_forest.fit(x_train,y_train)

    return modelo_random_forest

def ada_boost(seed, x_train, y_train) :
    from sklearn.ensemble import AdaBoostClassifier
    
    modelo_adaboost = AdaBoostClassifier(n_estimators=100, random_state=seed)

    modelo_adaboost.fit(x_train, y_train)

    return modelo_adaboost


````

# Questão 4

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados
- Treinar o modelo escolhido (`rf` ou `ab`)
- Avaliar o modelo
- Retornar a acurácia

**Solução**:

````python
def run_pipeline(model_type, seed):
    X_train, X_test, y_train, y_test = load_data()

    if model_type == "rf":
        modelo = random_forest(seed, X_train, y_train)
    elif model_type == "ad":
        modelo = ada_boost(seed, X_train, y_train)
    
    acuracia, predicao = evaluate(modelo, X_test, y_test)
    
    return modelo, acuracia, predicao
````

**Em qual profundidade começa o overfitting?**
**Por que a árvore consegue 100% no treino quando max_depth=None?**

1-  Ao analisar a curva de aprendizado, observa-se que o modelo atinge a acurácia máxima no treino rapidamente, mas mantém a estabilidade no conjunto de teste. A ausência de queda na acurácia de teste conforme a complexidade aumenta indica que o modelo não entrou em regime de overfitting, mas sim atingiu o seu limite de aprendizado para a arquitetura utilizada.

# Questão 8

Varie pelo menos um hiperparâmetro em cada modelo:

- Random Forest: `n_estimators`
- AdaBoost: `n_estimators`

## Analise:
- O desempenho muda significativamente?

## Responda:
- Qual modelo é mais sensível a mudanças?


O AdaBoost.
 O AdaBoost é um algoritmo de Boosting (aprendizado sequencial). Cada nova árvore é construída especificamente para corrigir os erros das anteriores, o que torna o modelo altamente dependente do número de iterações e da taxa de aprendizado.

Já o Random Forest utiliza Bagging (aprendizado paralelo e independente). Como as árvores não dependem umas das outras, o modelo é muito mais robusto e "estacionário", sendo pouco afetado por variações no número de estimadores após atingir a convergência inicial.

````python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier


rf_v2_model = RandomForestClassifier(n_estimators=200, random_state=42)
rf_v2_model.fit(X_train, y_train)


acc_rf_v2 = evaluate(rf_v2_model, X_test, y_test)


ad_v2_model = AdaBoostClassifier(n_estimators=200, random_state=42)
ad_v2_model.fit(X_train, y_train)


acc_ad_v2 = evaluate(ad_v2_model, X_test, y_test)

# --- RESULTADOS ---
print(f"RF (Original): {acc_rf:.4f} -> RF (200 est): {acc_rf_v2:.4f}")
print(f"AD (Original): {acc_ad:.4f} -> AD (200 est): {acc_ad_v2:.4f}")
````

1. A acurácia é suficiente para avaliar os modelos?
Não. Embora a acurácia indique o acerto global, ela pode ser enganosa se houver desequilíbrio entre as classes ou se o custo de erro for diferente para cada categoria (ex: confundir uma bota com um chinelo vs. uma camisa com um casaco). Como vimos nos resultados anteriores, o Random Forest teve uma acurácia alta, mas métricas de Precisão e Recall menores, o que revela que o modelo pode estar performando muito bem em algumas classes e falhando drasticamente em outras.

2. Como você garante que o resultado não ocorreu por acaso?
A principal forma de garantir a robustez é através da Validação Cruzada (Cross-Validation), onde o dataset é dividido em várias partes e o modelo é testado repetidamente em diferentes combinações. Outro ponto fundamental é a fixação do seed (random_state), que permite a reprodutibilidade dos experimentos, garantindo que a variação de desempenho não seja fruto de uma divisão "sortuda" entre os dados de treino e teste.

3. Cite dois possíveis problemas metodológicos neste experimento.
O primeiro problema é a falta de normalização ou padronização dos dados de entrada (pixels de 0 a 255); algoritmos de boosting costumam performar melhor com dados em escalas menores (0 a 1). O segundo é o uso de hiperparâmetros padrão (default); o AdaBoost, por exemplo, foi prejudicado por usar árvores rasas demais (stumps) para um problema de imagem complexo, o que não reflete o potencial real do algoritmo se ele fosse devidamente tunado.

4. O pipeline implementado é confiável? Justifique.
Parcialmente. Ele é confiável no sentido de seguir o fluxo padrão de Ciência de Dados (Carga -> Treino -> Avaliação) e garantir a separação entre treino e teste. No entanto, para ser considerado um pipeline de produção robusto, ele carece de etapas de limpeza de dados, busca exaustiva de parâmetros (GridSearch) e uma análise mais profunda de matriz de confusão, para entender exatamente onde os modelos estão errando antes de uma tomada de decisão final.
